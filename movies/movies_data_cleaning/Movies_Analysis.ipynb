## best & worst movies

```python
from IPython.display import HTML
pd.options.display.max_colwidth = 200
```

```python
# split and creat a new df_best from df 
df_best = df[["poster_path", "title", "budget_musd", "revenue_musd",
              "vote_count", "vote_average", "popularity"]].copy()
df_best
```

```python
# new cloumns profit_musd , return
df_best["profit_musd"] = df.revenue_musd.sub(df.budget_musd)
df_best["return"] = df.revenue_musd.div(df.budget_musd)
```

```python
# just to see that the new cloumns are made
df_best
```

```python
# change the columns names and  (ROI = Return On Investment)
df_best.columns = ["", "Title", "Budget", "Revenue", "Votes", 
                   "Average Rating", "Popularity", "Profit", "ROI"]
```

```python
# set the main index by the Title
df_best.set_index("Title", inplace = True)
```

```python
df_best
```

```python
#df_best.iloc[0,0]
```

```python
#subset = df_best.iloc[:5, :2]
#subset
```

```python
#HTML(subset.to_html(escape=False))
```

```python
#  get highest rat movie 
df_best.sort_values(by = "Average Rating", ascending = False)
# the prob here is we dont have lots of vote on the top rat movies 
```

```python
# addign 1 more filter 
df_best.sort_values(by = "ROI", ascending = False)
# here the prob is that the top movies we filer have 0 budget
```

```python
# set the min budget for 5mill 
df_best.loc[df_best.Budget >= 5].sort_values(by = "ROI", ascending = False)
```

```python
# filling the Budget and Votes nans with 0 
df_best.Budget.fillna(0, inplace = True)
df_best.Votes.fillna(0, inplace = True)
```

```python
df_best.info()
```

```python
# creat function that helps to find best_worst
def best_worst(n, by, ascending = False, min_bud = 0, min_votes = 0):
    
    
    df2 = df_best.loc[(df_best.Budget >= min_bud) & (df_best.Votes >= min_votes), 
                      ["", by]].sort_values(by = by, ascending = ascending).head(n).copy()
    
    return HTML(df2.to_html(escape=False))
```

```python
# top movies - highest Revenue
best_worst(n = 5, by = "Revenue")
```

```python
# top movies - highest Budget
best_worst(5, "Budget")
```

```python
# top movies - highest Profit
best_worst(5, "Profit")
```

```python
# top movies - lowest profit
best_worst(5, "Profit", ascending = True)
```

```python
# top movies - highest ROI
best_worst(5, "ROI", min_bud = 50)
```

```python
# top movies - lowest ROI
best_worst(5, "ROI", ascending = True, min_bud = 100)
```

```python
# top movies - most votes
best_worst(5, "Votes")
```

```python
# top movies - highest rating
best_worst(5, "Average Rating", min_votes = 50)
```

```python
# top movies - lowest rating
best_worst(5, "Average Rating", ascending = True, min_votes = 100)
```

```python
# here we see the lowest rating with extra conditions 
best_worst(5, "Average Rating", ascending = True, min_votes = 20, min_bud = 20)
```

```python
# most popular
best_worst(5, "Popularity")
```

## common words in movie titles & taglines

```python
# word cloud is a data visualization technique used for representing text data in which the size of each word indicates its frequency
# or importance. Significant textual data points can be highlighted using a word cloud. Word clouds are widely used for analyzing 
# data from social network websites.
from wordcloud import WordCloud
```

```python
df
```

```python
df.tagline[1:2]
```

```python
df.overview[1:2]
```

```python
# drop nan
# stage 1 for the WordCloud
title = df.title.dropna()
overview = df.overview.dropna()
tagline = df.tagline.dropna()
```

```python
title
```

```python
# make 1 string with space between the words
# stage 2 for WordCloud
' '.join(title)
```

```python
# stage 3 for WordCloud
title_corpus = ' '.join(title)
overview_corpus = ' '.join(overview)
tagline_corpus = ' '.join(tagline)
```

```python
tagline_corpus
```

```python
# find and vis the most com word in 1 string and creat img
title_wordcloud = WordCloud(background_color='white', height=2000, width=4000, max_words= 200).generate(title_corpus)
title_wordcloud
```

```python
plt.figure(figsize=(16,8))
plt.imshow(title_wordcloud, interpolation= "bilinear")
plt.axis('off')
plt.show()
```

```python
tagline_wordcloud = WordCloud(background_color='white', height=2000, width=4000).generate(tagline_corpus)
plt.figure(figsize=(16,8))
plt.imshow(tagline_wordcloud, interpolation= "bilinear")
plt.axis('off')
plt.show()
```

```python
overview_wordcloud = WordCloud(background_color='white', height=2000, width=4000).generate(overview_corpus)
plt.figure(figsize=(16,8))
plt.imshow(overview_wordcloud, interpolation= "bilinear")
plt.axis('off')
plt.show()
```
